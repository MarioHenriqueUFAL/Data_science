Atributos categóricos - LabelEncoder

```python
#Na base de dados temos variáveis categóricas, por exemplo, a coluna "workclass", "relationship"
# E temos variáveis numéricas
#Para as variáveis categóricas, precisamos converter essas strings em números, de modo que os processos
#de aprendizado de máquina possam ser realizados. Pois o aprendizado de máquina envolve operçaões matemáticas
#dai a importância de converter as variáveis categóricas em forma numérica
#Para fazer essa transformação usamos o "LabelEncoder"
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_workclass = LabelEncoder() #1
label_encoder_education = LabelEncoder() #3
label_encoder_matiralstatus = LabelEncoder() #5
label_encoder_occupation = LabelEncoder() #6
label_encoder_relatioship = LabelEncoder() #7
label_encoder_race = LabelEncoder() #8
label_encoder_sex = LabelEncoder() #9
label_encoder_nativecountry = LabelEncoder() #13
```

```python
#Tranformando os valores categóricos em numéricos com o LabelEncoder
x_census[:,1] = label_encoder_workclass.fit_transform(x_census[:,1])
x_census[:,3] = label_encoder_education.fit_transform(x_census[:,3])
x_census[:,5] = label_encoder_matiralstatus.fit_transform(x_census[:,5])
x_census[:,6] = label_encoder_occupation.fit_transform(x_census[:,6])
x_census[:,7] = label_encoder_relatioship.fit_transform(x_census[:,7])
x_census[:,8] = label_encoder_race.fit_transform(x_census[:,8])
x_census[:,9] = label_encoder_sex.fit_transform(x_census[:,9])
x_census[:,13] = label_encoder_nativecountry.fit_transform(x_census[:,13])
```

```python
#Note que os atributos previsores são todos numéricos, prontos para as técnicas de machine learning
#Como podemos perceber no resultado abaixo, temos várias colunas com escalas diferentes. 
#Os algoritmos de aprendizado de máquina interpretam as escalas diferentes por peso
#Ou seja, uma coluna com dados numéricos maiores, em valor absoluto, diz ao algoritmo que esse dado é mais importantes
#que os demais. Para solucionar esse problemas, temos que trabalhar com o OneHotEncoder
#O label encoder transforma as variáveis categóricas em variáveis numéricas, mas, por exemplo, para uma coluna
#de carros que podem ter três tipos de carros: gol, chevette, monza; o LabelEnconder trasnforma esses dados
#em numérico. Logo, teríamos as seguintes representações: (1) gol, (2) chevette, (3) monza. Porém,
#Os algoritmos de aprendizado de máquina podem interpretar que o monza é mais importante que os demais,
#uma vez que apresenta a representação numérica com maior valor absoluto. Para corrigir isso, usamos o OneHotEncoder
#O OneHotEncoder transforma a representação numérica em representação do tipo "Dummy",isto é, em representação
#Usando números binários. Usando o OneHotEncoder, teríamos a seguinte representação para a coluna carros:
# gol (1,0,0)
#chevette (0,1,0)
#monza(0,0,1)
#Dessa forma, eliminamos o problema anterior. Pois não há diferentes pesos para esses carros, são variáveis
#categóricas do tipo nominal e não ordinal.
x_census[0]
```

Trabalhando com o #OneHotEncoder

```python
#Primeiramente trabalhamos com o "LabelEncoder" para transformar as variáveis categóricas em variáveis numéricas
#Em seguida, trabalhamos com o OneHotEncoder para transformar os dados em uma representação binária, uma nova codificação
#Procedimento clássico:
#Transformar as variáveis categóricas nominais em variáveis numéricas usando o LabelEncoder
#Transformar as variáveis numéricas, criadas com o LabelEncoder, em variáveis dummy usando o OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Esse comando vai realizar uma nova codificação. Por exemplo, na coluna "workclass" temos 9 valores categóricos possíveis
#Inicialmente, temos uma única coluna, onde cada valor possível é identificado por um número diferente
#No entanto, os números são interpretados pelo algoritmo de machine learning como pesos
#Usando esse comando, a coluna Workclass vai conter 9 colunas, cada valor possível é representado por
#numeração binária, de modo a eliminar o problema anteriormente criado.
#O parâmetro "remainder = 'passthroug'" deve ser inserido para que as demais colunas não sejam apagadas ou alteradas
#Note que as colunas a serem transformadas foram passadas como uma lista de índices
onehotencoder_census = ColumnTransformer(transformers = [('OneHot', OneHotEncoder(),[1,3,5,6,7,8,9,13])], remainder = "passthrough")
```

```python
#redefinindo a base de dados, temos:
x_census = onehotencoder_census.fit_transform(x_census).toarray()
```

```python
x_census
```

```python
#É possível observar que temos novas colunas que foram criadas, isso é devido a nova representação binária
#criada a partir do OneHotEncoder
x_census[0]
```
