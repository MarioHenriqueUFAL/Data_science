**Base credit data**

```python
#abrindo o arquivo pré-tratado
with open('credit.pkl','rb') as f:
  x_credit_treinamento,y_credit_treinamento,x_credit_teste,y_credit_teste = pickle.load(f)
```

```python
x_credit_teste.shape, y_credit_teste.shape
```

```python
x_credit_treinamento
```

```python
y_credit_treinamento
```

```python
x_credit_teste
```

```python
y_credit_teste
```

```python
#Importando a biblioteca para o algoritmo de árvore de decisão e realizando o treinamento do algoritmo
#com os dados de treinamento "x_credit_treinamento" e "y_credit_treinamento", temos:
from sklearn.tree import DecisionTreeClassifier
arvore_credit = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
arvore_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
#testando o algoritmo com os registros de teste, temos:
previsoes = arvore_credit.predict(x_credit_teste)
```

```python
previsoes
```

```python
#comparando os resultados previstos com os resultados reais da base de dados, temos:
from sklearn.metrics import accuracy_score, classification_report
```

```python
#a função accuracy_score(classes verdadeiras, classes previstas) é utilizado para prever a eficácia geral do
#algoritmo. No entanto, quando temos dados desbalanceados (mais dados de uma classe que da outra), tal valor
#pode ser mascarado para uma das classes. Ou seja, uma classe pode estar sendo prevista com boa eficácia
#enquanto a outra (com menos dados) está sendo prevista com eficiência muito ruim. Para resolver isso,
#podemos usar a função "classification_report"
print(accuracy_score(y_credit_teste,previsoes))
```

```python
accuracy = accuracy_score(y_credit_teste,previsoes)
accuracy
```

```python
#gerando a matriz de confusão:
#O eixo y da matriz de confusão indica os valores verdadeiros, enquanto o eixo x os valores previstos 
from yellowbrick.classifier import ConfusionMatrix
cn = ConfusionMatrix(arvore_credit)
cn.fit(x_credit_treinamento, y_credit_treinamento)
cn.score(x_credit_teste,y_credit_teste)
```

```python
print(classification_report(y_credit_teste,previsoes))
#recal: do conjunto de dados reais, o quanto de dados desse conjunto o algoritmo identifica corretamento
```

```python
#Podemos visualizar a árvore de decisão criada
from sklearn import tree
previsores = ['income','age','loan']
fig, axes = plt.subplots(nrows=1,ncols=1, figsize =(20,20))
tree.plot_tree(arvore_credit,feature_names = previsores, class_names = str(arvore_credit.classes_),filled=True);
fig.savefig('arvore_credit.png')
```

Random Forest (florestas randômicas)

Base credit data

```python
#random forest é uma melhoria da árvore de decisão. Na prática ela utiliza várias árvores de decisão
#em conjunto para obter a média (regressão) e maioria de classificação (classificação)
#É baseada em aprendizagem em conjuntom, é como se estivéssemos consultando vários especialistas
#para a tomada de decisões.
#Para aplicar o algoritmo de Random Forest, procedemos da seguinte forma:
from sklearn.ensemble import RandomForestClassifier
```

```python
import pickle
with open('credit.pkl','rb') as f:
  x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
  
```

```python
x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
x_credit_teste.shape, y_credit_teste.shape
```

```python
#treinando o algoritmo de aprendizado de máquina Random Forest, temos:
#A função "RandomForestClassifier" recebe três argumentos neste exemplo, os quais são definidos como:
#n_estimators define a quantidade de árvores de decisão que serão criadas no algoritmo. Lembrando que, se temos
#10 árvores de decisão e 7 árvores classificam o risco como alto e 3 classificam como baixo, pelo voto da maioria
#a classificação retornada pelo algoritmo é que o risco é alto.
random_forest_credit = RandomForestClassifier(n_estimators = 40, criterion ='entropy',random_state = 0)
```

```python
#realizando o treinamento do algoritmo com o conjunto de dados de treinamento, temos:
random_forest_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = random_forest_credit.predict(x_credit_teste)
```

```python
#Após o treinamento do algoritmo de random foreste, precisamos comparar os dados previstos pelo algoritmo
#com os dados reais da base de dados histórica
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_credit_teste,previsoes))
```

```python
#O eixo y representa os dados da base de dados histórica (verdadeiros)
#O eixo x representa os dados previstos pelo algoritmo
from yellowbrick.classifier import ConfusionMatrix
cn = ConfusionMatrix(random_forest_credit)
cn.fit(x_credit_treinamento,y_credit_treinamento)
cn.score(x_credit_teste,y_credit_teste)
```

```python
print(classification_report(y_credit_teste,previsoes))
```
